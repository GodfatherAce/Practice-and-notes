# Seaborn
## Distplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
print recent_grads.columns

# kde argument is for turning off kernal densitiy estimate plotting
sns.distplot(recent_grads['Median'], kde=False)
sns.plt.show()
```

**Seaborn can be used only to plot things more nicely, if desired, by using sns.plt.show() and everything left untouched**

```python
recent_grads['Median'].hist()
sns.plt.show()
```

### Customizing distplot

```python
sns.distplot(recent_grads['Median'], kde=False)
sns.axlabel('Median', 'Frequency')
sns.plt.show()
```

```python
sns.set_style(style='dark')
sns.distplot(recent_grads['Median'], kde=False)
sns.axlabel('Median', 'Frequency')
sns.plt.show()
```

## Boxplot

```python
sns.boxplot(x=recent_grads['Major'],y=recent_grads['Median'])
sns.plt.show()
```

## Pairplot

comparing n variables with itself and others. nxn plots
- versus itself is a histogram
- versus others is a scatterplot

```python
columns=['Median','Total','Rank']
sns.pairplot(recent_grads[columns])
sns.plt.show()
```
